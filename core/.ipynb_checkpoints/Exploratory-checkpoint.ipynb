# Model Output Components

---

### 1. 'contours' (Refined Final Contours)

* **Content:** A list (of batch size, 1 in this case) of tensors representing the coordinates of the **final contours** for each detected object.
* **Shape (per object):** (Number of Points, 2). Each row is a coordinate $[Y, X]$.
* **Meaning:** The coordinates of the boundaries of the detected cells **after refinement**. This is the **most accurate representation** of the cell's shape.
* **Napari Visualization:** Corresponds to the **Shapes layer** or **Labels layer**.

---

### 2. 'boxes' (Bounding Boxes)

* **Content:** A list of tensors representing the **bounding boxes** of the final detections.
* **Shape (per object):** (Number of Detections, 4). Each row is $[Y_{min}, X_{min}, Y_{max}, X_{max}]$.
* **Meaning:** The rectangles that enclose each detected cell, used primarily for **Non-Maximum Suppression (NMS)**.

---

### 3. 'scores' (Confidence Scores)

* **Content:** A list of tensors representing the **confidence score** (probability) for each bounding box.
* **Shape (per object):** (Number of Detections,). A floating-point value per detection.
* **Meaning:** Indicates the model's **certainty**. Values are between $0.0$ and $1.0$.

---

### 4. 'classes' (Detected Classes)

* **Content:** A list of tensors indicating the **class** to which each detection belongs.
* **Shape (per object):** (Number of Detections,). An integer value per detection.
* **Meaning:** For cell segmentation, this is a **single-class model**. All values are typically **1** ("cell").

---

### 5. 'locations' (Detection Center Locations)

* **Content:** A list of tensors representing the **central location** of each detection.
* **Shape (per object):** (Number of Detections, 2). Each row is the coordinate $[Y_{center}, X_{center}]$.
* **Meaning:** The central points used by the model as a starting point for contour proposal.

---

### 6. 'fourier' (Fourier Coefficients)

* **Content:** A list of tensors representing the **Fourier Transform coefficients** for the contours.
* **Shape (per object):** (Number of Detections, N, 4).
* **Meaning:** The mathematical representation of the **initial contour proposal** (using Fourier expansion) before refinement.

---

### 7. 'contour_proposals' (Initial Contour Proposals)

* **Content:** A list of tensors with the coordinates of the **initial proposed contours**.
* **Meaning:** The shape prediction **before** the model applies the refinement stage. The difference with `'contours'` shows the improvement from refinement.

---

### 8. 'box_uncertainties' (Box Uncertainty)

* **Content:** A list of tensors representing the **estimated uncertainty** of each bounding box coordinate.
* **Shape (per object):** (Number of Detections, 4). Each row is [Ymin Uncertainty, Xmin Uncertainty, Ymax Uncertainty, Xmax Uncertainty].
* **Meaning:** A measure of the model's **reliability** in the precision of the box coordinates. Lower values mean higher certainty.

'boxes' -> [Y_min, X_min, Y_max, X_max]

'scores' -> punctuation per bbox

'locations' -> [Y_centro, X_centro] 

'box_uncertainties' -> [Incertidumbre Ymin, Incertidumbre Xmin, Incertidumbre Ymax, Incertidumbre Xmax]

```python
from bioio import BioImage
import os
import numpy as np 
import bioio_tifffile
from tqdm import tqdm
import matplotlib.pyplot as plt
import celldetection as cd
import torch
import skimage.io
import cv2
from bioio.writers import OmeTiffWriter
from skimage.draw import polygon
import shutil
import random
import math
```

```python
def Percentile_Normalization(projection:np.array, percentile_low: float =0.00,percentile_high: float =100.00):    
    
    low_val = np.percentile(projection, percentile_low)
    high_val = np.percentile(projection, percentile_high)
    if high_val <= low_val:
        max_val = np.max(projection)
        if max_val > 0:
            normalization = projection / max_val
        else:
            normalization = projection
    else:
        normalization = (projection - low_val) / (high_val - low_val)
    normalization = np.clip(normalization, 0.0, 1.0)

    return normalization

def Format_chanels(projection:np.array):
    if len(projection.shape) == 2:
        projection = np.stack([projection] * 3, axis=-1).astype(np.float32)
    if projection.shape[0] == 1:
        projection = np.repeat(projection, 3, axis=0)
    return projection

def Charge_model():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model_name = 'ginoro_CpnResNeXt101UNet-fbe875f1a3e5ce2c'
    model = cd.fetch_model(model_name, check_hash=True).to(device)
    return model.eval(), device
    
    
def Predictor(model, device, scene : np.array, mood:str = 'both'):
    
     
    order = 2 # Contour order. The higher, the more complex contours can be proposed. order=1 restricts the CPN to propose ellipses, order=3 allows for non-convex rough outlines, order=8 allows even finer detail.
    nms_thresh = 0.31 #  IoU threshold for non-maximum suppression (NMS). NMS considers all objects with iou > nms_thresh to be identical.
    score_thresh = 0.90 #Score threshold. For binary classification problems (object vs. background) an object must have score > score_thresh to be proposed as a result.
    samples = 224.0 #sampling points  Number of samples. This sets the number of coordinates with which a contour is defined. This setting can be changed on the fly, e.g. small for training and large for inference. Small settings reduces computational costs, while larger settings capture more detail.
    refinement = False # Whether to use local refinement or not. Local refinement generally improves pixel precision of the proposed contours.
    refinement_iterations = 4.0 #refinement step  Number of refinement iterations.
    refinement_margin = 700.0 #tile size Maximum refinement margin (step size) per iteration.
    refinement_buckets =  300.0 #overlap size Number of refinement buckets. Bucketed refinement is especially recommended for data with overlapping objects.

    if mood == 'max':
        projection = np.max(scene, axis=0).astype(np.float32)
    if mood == 'std':
        projection = np.std(scene, axis=0).astype(np.float32)
    if mood == 'both':
        projection = (np.max(scene, axis=0)+np.std(scene, axis=0)).astype(np.float32)

    projection = Percentile_Normalization(projection=projection)
    
    projection = Format_chanels(projection=projection)

    in_channels = projection.shape[0]
    
    with torch.no_grad():
        x = cd.to_tensor(projection, transpose=True, device=device, dtype=torch.float32)
        x = x[None]
        y = model(x, in_channels=in_channels, order=order, score_thresh=score_thresh, samples=samples, refinement=refinement, refinement_iterations=refinement_iterations, refinement_margin=refinement_margin, refinement_buckets=refinement_buckets)
    return projection , y

def Plot_result(projection,bboxes,contours):
    #this commented lines generate a binary mask based on the model segmentation
    # skimage_contours = []
    # for tensor_contour in y['contours'][0].cpu().numpy():
    #     contour_np = tensor_contour.astype(np.int32)
    #     skimage_contours.append(contour_np)
    # binary_mask = np.zeros((scene.shape[1],scene.shape[2]), dtype=np.uint8)
    # for contour in skimage_contours:
    #     r = contour[:, 1] 
    #     c = contour[:, 0]
    #     rr, cc = polygon(r, c, shape=(scene.shape[1],scene.shape[2]))
    #     binary_mask[rr, cc] = 1
        
    plt.figure(figsize=(10, 10))
    plt.imshow(projection) 
    plt.axis('off')
    cd.plot_boxes(bboxes)
    cd.plot_contours(contours)
    # plt.imshow(binary_mask, cmap='gray')

def move_directories(list_subdirs, destination):
    os.makedirs(destination, exist_ok=True)
    for subdir_name in list_subdirs:
        source_path = os.path.join(SOURCE_DIR, subdir_name)
        destination_path = os.path.join(destination, subdir_name)
        
        try:
            shutil.move(source_path, destination_path)
        except Exception as e:
            print(f"Error moving {subdir_name} to {destination}: {e}")
 
```
