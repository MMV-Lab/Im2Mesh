```python
import numpy as np
from cellpose import models, core, io, plot, utils
from pathlib import Path
from tqdm import trange
import matplotlib.pyplot as plt
from natsort import natsorted

io.logger_setup() # run this to get printing of progress

#Check if colab notebook instance has GPU access
if core.use_gpu()==False:
  raise ImportError("No GPU access, change your runtime")

model = models.CellposeModel(gpu=True)
```

```python
dir = "/mnt/eternus/users/Jair/Im2Mesh_Myversion/100x_multicell/"
dir = Path(dir)
if not dir.exists():
  raise FileNotFoundError("directory does not exist")

# *** change to your image extension ***
image_ext = ".tif"

# list all files
files = natsorted([f for f in dir.glob("*"+image_ext) if "_masks" not in f.name and "_flows" not in f.name])

if(len(files)==0):
  raise FileNotFoundError("no image files found, did you specify the correct folder and extension?")
else:
  print(f"{len(files)} images in folder:")

for f in files:
  print(f.name)
```

```python
norm = 'none' # 'min-max' 'standar' 'none'
masks_ext = ".png" if image_ext == ".png" else ".tif"
for i in trange(len(files)):
    f = files[i]
    img = io.imread(f)
    if norm != 'none':
        if norm == 'min-max':
            img_min = img.min()
            img_max = img.max()
            img = (img - img_min) / (img_max - img_min)
        if norm == 'standar':
            mean = np.mean(img)
            std = np.std(img)
            img = (img - mean) / std
    print(f'your image has shape: {img.shape}')
    masks, flows, styles = model.eval(img, z_axis=0, channel_axis=3,batch_size=32,do_3D=True, flow3D_smooth=2)
    io.imsave( '/mnt/eternus/users/Jair/Im2Mesh_Myversion/100x_multicell_segmentations_cellpose/' + (f.stem + masks_ext), masks)
```
